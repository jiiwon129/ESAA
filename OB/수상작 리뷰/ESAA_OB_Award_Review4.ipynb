# **새만금 개발청 공공 빅데이터 활용 공모전**

새만금 개발청
- 전라북도 군산, 김제, 부안 일대에 위치한 새만금 지역을 종합적으로 관리 및 개발하기 위해 설립된 중앙행정기관

## **주제:** 새만금의 재난대응 역량 강화를 위한 방재공원 최적 입지분석

- 최우수상 (김수로팀) 수상작 참고

## **선정배경**

개발이 현재 진행되고 있는 새만금은 다음에 대한 준비 필요
1. 배수로 및 침수 대책 방안
2. 2차 전지 관련 기업 입주 예정으로 화재 예방 방안
3. 녹지 공간 활용 방안

기후 변화 및 도시화 -> 자연 재난 유형의 복잡화 및 재난 규모 대형화

[결론]
재난 대응 역량 강화 및 시민의 안전을 도모하기 위한 방재공원 설치
1. 방재공원 내 빗물 저류 배수시설로 침수 피해 예방
2. 공원용지로 계획된 위치 활용 -> 녹지 공간 활용성 증대


\* 방재공원: 재해로부터 안전을 지키기 위한 기능을 갖춘 공원
  - 평상시: 일반 공원처럼 산책로, 운동 시설, 녹지 공간 제공
  - 재난 발생 시: 넓은 광장과 임시 천막 설치 공간 제공 등 대피 공간, 임시 거주지, 구호 거점과 같은 역할을 함


## **분석 개요**

1. 예상 수요 및 접근성 반영 -> 방재공원으로 적절한 입지 후보 산출
2. 최종 도출된 입지 근처의 공원용지로 계획된 곳을 방재공원 목적으로 설립 제안

## **분석 활용 기법**

### **QGIS 공간 분석**

image.png

- 전북 군산, 김제, 부안 일대(새만금 지역)를 배경으로 한 공간 데이터 시각화  

1. **지오코딩 기반 공간 정보화**  
   - 주소 및 좌표 데이터를 지도상 위치로 변환하여 새만금 지역 데이터를 공간적으로 시각화  

2. **새만금 지역 정보 QGIS 구현**
   - 오픈소스 GIS 툴(QGIS)을 활용하여 새만금 관련 데이터를 지도에 추가하여 분석 및 시각화  

3. **인구 및 건축물 밀집도 분석**  
   - 새만금 지역의 인구 분포 및 건축물 밀집도를 지도 기반으로 파악  
   - 특정 지역의 개발 정도, 활용도, 향후 개발 가능성을 분석하는 데 활용 가능

#### **QGIS란?**

- **QGIS(Quantum GIS)**는 오픈소스 기반의 **무료 지리정보시스템(GIS) 소프트웨어**  
- 지리공간 데이터를 **수집, 저장, 분석, 시각화** 가능  
- 다양한 분야(행정, 도시계획, 환경, 교통, 개발사업 등)에서 활용  

[주요 기능]
1. **지오코딩**: 주소·좌표 데이터를 지도상 위치로 변환  
2. **레이어 분석**: 행정구역, 인구, 도로, 건축물 등 여러 공간정보를 겹쳐서 분석  
3. **시각화**: 지도 위에 데이터 분포를 직관적으로 표현

## **차별점 및 배울점**

### 1. 공간 분석 기술의 적용

단순한 통계 분석을 넘어, 데이터의 위치 정보를 활용하여 지리적 맥락 파악 및 시각화하는 능력을 잘 보여줌
- QGIS와 같은 GIS 툴을 이용하여 공간 데이터를 시각화하고, 인구 밀집도 분석, 접근성 분석 등 실제 공간 분석 기법을 적용한 점이 방재공원 입지 선정에 적합한 방법론이라고 판단됨.


### 2. 분석 결과에 대한 비판적 사고 및 현실적 제안

데이터 분석 결과(ex. 군집 중심점)를 맹신하지 않고, 현실적인 제약 사항을 고려하여 최종 결론을 도출함
- 후보지 1의 경우, 분석 결과는 최적의 위치를 가리켰지만, 토지 이용 계획상 산업용지라는 사실을 파악하고 인근 공원 용지를 대안으로 제시

--> 분석 결과가 항상 현실과 일치하지 않을 수 있음을 인식 및 데이터 외부 정보(ex. 토지 이용 계획)을 결합하여 분석의 타당성을 검증 및 보완하는 비판적 사고를 보임

### 3. 시각화 및 커뮤니케이션 역량

QGIS를 활용한 분석 결과 시각화는 프로젝트의 이해도를 높이는 데 많은 도움을 줌.
지도로 표현된 인구 밀집도나 후보지 위치는 복잡한 데이터를 직관적으로 전달하는 강력한 수단이 되므로 복잡한 분석 결과를 비전문가도 쉽게 이해할 수 있도록 효과적으로 시각화하고, 분석 과정과 결론을 명확하게 설명함.
