**데이터 셀렉션 및 필터링**

---

DataFrame의 [ ] 연산자

- pandas와 numpy의 데이터 셀렉션과 필터링은 혼동하기 쉽기 때문에 유의해야함

 - numpy : '[ ]' 연산자 내 단일 값 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱을 통해 데이터 추출
 - pandas : iloc[ ], loc[ ] 연산자를 통해 데이터 추출

- DataFrame 뒤 [ ]는 col만 지정할 수 있는 'col 지정 연산자'로 이해하는 것이 좋음
 - 즉, DataFrame 뒤 [ ] 연산자에는 숫자형이 아닌 칼럼명/칼럼의 리스트 지정
 - 슬라이싱을 이용한 데이터 추출은 추천하지 않음

DataFrame iloc[ ] 연산자

- iloc[ ] : 위치(Location) 기반 인덱싱 방식

 - 행과 열의 위치, 0을 출발점으로 하는 세로축, 가로축 좌표 정수값으로 지정

 - iloc[ ] 연산자 내 '위치 정수값', '위치 정수 슬라이싱'을 이용하여 데이터 추출

 - 위치 기반 인덱싱'만' 허용

- pandas 인덱싱에서의 '-1'

 - 맨 마지막 데이터 값을 의미 (numpy와 동일)

 iloc[:, -1] : 맨 마지막 칼럼의 값 ( == 타깃값)

 iloc[:, :-1] : 맨 마지막 칼럼을 제외한 모든 칼럼의 값 ( == 피처값)

- 슬라이싱, 팬시 인덱싱 제공

- 불린 인덱싱 미제공
 - 명확한 위치 기반 인덱싱이 사용되어야 하는 제약 때문

DataFrame loc[ ] 연산자

- loc[ ] : 명칭(Label) 기반 인덱싱 방식

 - 행 위치에 DataFrame 인덱스 값, 열 위치에 칼럼명 입력

   -> loc[ 인덱스값, 칼럼명 ]

- 슬라이싱 기호 ':' 적용 시 유의
 - 일반적인 슬라이싱 같은 경우

   '시작값 : 종료값' -> 시작 값 ~ '종료 값-1'

 - loc[ ]에서의 슬라이싱

   '시작값 : 종료값' -> 시작 값 ~ '종료 값'

    -> 명칭은 숫자형이 아닐 수 있기 때문에 -1을 할 수 없기 때문

불린 인덱싱

- [ ], loc[ ] 지원
- iloc[ ] 미지원

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
```

```python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60
                                                                                else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
  cat=''
  if age<=5: cat='Baby'
  elif age<=12: cat='Child'
  elif age<=18: cat='Teenager'
  elif age<=25: cat='Student'
  elif age<=35: cat='Young Adult'
  elif age<=60: cat='Adult'
  else: cat='Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head()
```
